```python
import numpy as np
#importer le module pyplot du package matplotlib
import matplotlib.pyplot as plt
#Création de la fonction
x=np.linspace(0,20,60)
y=x*x

#Tracer le graphe en points
plt.scatter(x,y,c='red',lw=2,marker='+')
#Tracer le graphe en ligne continue
plt.plot(x,y,c='blue',lw=1)
#Afficher le graphe
plt.show() #C'est optionnel sur Jupyter

#Sauvegarder la figure
plt.savefig("Plot generated using Matplotlib.png") #choisisez l'extension que vous voulez (.jpg.png.....)
```

```python
def rampe(t): #Définir la fonction rampe
    if t > 0:
        return t
    else :
        return 0
    
def echelon(t): #Définir la fonction échelon
    if t>=0 :
        return 1
    else :
        return 0
#Cycle de vie d'un graphe sur Matplotlib
import matplotlib.pyplot as plt             
import numpy as np

x=np.linspace(-1 , 1, 100)
x=np.linspace(-10,10,100)

plt.figure(figsize=(12,3))
#Sous figure 1
plt.subplot(1,2,1)
y=[echelon(t) for t in x]
plt.plot(x , y ,label="échelon" , c="red",lw=3)
plt.title("La fonction échelon")
plt.xlabel('Temps')
plt.legend()

#Sous figure 2
plt.subplot(1,2,2)
y=[rampe(t) for t in x]
plt.title("La fonction rampe")
plt.scatter(x , y ,label='rampe', c='orange',lw = 2)
plt.xlabel('Temps')
plt.legend()
```
