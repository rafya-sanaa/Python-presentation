```python
import numpy as np
#importer le module pyplot du package matplotlib
import matplotlib.pyplot as plt
#Création de la fonction
x=np.linspace(0,20,60)
y=np.sin(x)

#Tracer le graphe en points
plt.scatter(x,y,c='red',lw=2,marker='+')
#Tracer le graphe en ligne continue
plt.plot(x,y,c='blue',lw=1)
#Afficher le graphe
plt.show() #C'est optionnel sur Jupyter
```

```python
#importer le module pyplot du package matplotlib
import matplotlib.pyplot as plt
#Création de la fonction
x=np.linspace(0,20,60)
y=np.sin(x)

#Tracer le graphe dans la première subdivision de figure
plt.subplot(1,2,1)
plt.scatter(x,y,c='orange',lw=2,marker='+')

#Tracer le graphe dans la deuxième subdivision de figure
plt.subplot(1,2,2)
plt.plot(x,y,c='blue',lw=1)
```

```python
#Créer une nouvelle figure de largeur 12 et de hauteur 3 en pouces
plt.figure(figsize=(12,3))

#Sous-figure1
plt.subplot(1,2,1)
plt.scatter(x,y,c='orange',lw=2,marker='+')

#Sous-figure2
plt.subplot(1,2,2)
plt.plot(x,y,c='blue',lw=1)
```

```python
x=np.linspace(0,20,60)
#Création des graphes
plt.plot(x,np.sin(x),label="sin(x)",c='red',lw=2)
plt.plot(x,np.cos(x),label="cos(x)",c='black',lw=3)
 #Ajouter un titre
plt.title("Signaux sinusoidaux")
#Nommer les axes
plt.xlabel('abscisses')
plt.ylabel('ordonnées')
#Ajouter une légende مفتاح
plt.legend()
```
