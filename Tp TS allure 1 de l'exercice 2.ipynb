```python
#Définir la fonction rampe
def rampe(t):
    if t>0 :
        return t
    else :  
        return 0
    
#Définir la fonction X(t)
def X(t):
    return -rampe(t-3)

import matplotlib.pyplot as plt
import numpy as np

x= np.linspace(-1,10,150)
y = [X(t) for t in x]


plt.plot(x , y , label="x(t)",ls='--',c="black")

plt.xlabel("t en ms")
plt.legend()
```

```python
#Définir la fonction X(t)
def X(t):
    return rampe(t)-rampe(t-3)


x= np.linspace(-1,10,150)
y = [X(t) for t in x]


plt.plot(x , y , label="x(t)",ls='--',c="black")

plt.xlabel("t en ms")
plt.legend()
```
