```python
#Allure 1 

#Définir la fonction rampe
def rampe(t):
    if t>0 :
        return t
    else :  
        return 0
    
#Définir la fonction X(t)
def X(t):
    return rampe(t)-2*rampe(t-3)+rampe(t-4)-rampe(t-6)+rampe(t-8)

import matplotlib.pyplot as plt
import numpy as np

x= np.linspace(-1,10,150)
y = [X(t) for t in x]
z = [X(-4*t+8) for t in x]

plt.plot(x , y , label="x(t)",ls='--',c="black")
plt.plot(x , z , label="y(t)",lw=3)
plt.xlabel("t en ms")
plt.legend()

```

```python
#Allure 3 Méthode 1
    
#Définir la fonction porte
def Porte1(t):
    if (t>-1/2 and t<1/2):
        return 1
    else:
        return 0 

import matplotlib.pyplot as plt
import numpy as np

x1= np.linspace(-3,3,150)
y = [Porte1(t) for t in x1]

#Créer une figure
plt.figure(figsize=(10,2))

plt.plot(x1 , y , label="Porte",c="red")
plt.xlabel("t en ms")
plt.title("La fonction porte")
plt.legend()
```

```python
#Allure 3 Méthode 2

#Définir la fonction échelon
def echelon(t): 
    if t>=0 :
        return 1
    else :
        return 0
    
#Définir la fonction porte
def Porte2(t):
    return echelon(t+1/2)-echelon(t-1/2)

import matplotlib.pyplot as plt
import numpy as np

x1= np.linspace(-3,3,150)
y = [Porte2(t) for t in x1]

#Créer une figure
plt.figure(figsize=(10,2))

plt.plot(x1 , y , label="Porte",c="red")
plt.xlabel("t en ms")
plt.title("La fonction porte")
plt.legend()
```
